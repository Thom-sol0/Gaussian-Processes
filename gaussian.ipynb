```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rnd
```

```python
values_set = np.arange(0,6,0.01)
n_values = len(values_set)

n_samples = 10
```

```python
def K(s,t):
    return np.exp((1+np.abs(s-t))*np.exp(-np.abs(s-t)))

def M(i,j):
    return K(values_set[int(i)],values_set[int(j)])
```

```python
# gaussian vector
mean_vector = np.zeros(n_values)
matrix = np.zeros((n_values,n_values))
for i in range(n_values):
    for j in range(n_values):
        matrix[i,j] = M(i,j)


# create 1000 samples of gaussian vector with mean_vector and cov_matrix
X = rnd.multivariate_normal(mean_vector, matrix, n_samples)
print( matrix )
```

```python
for j in range(n_samples):
    plt.plot(values_set, X[j], color='r', alpha=1)
```
